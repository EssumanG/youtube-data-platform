```python
import pandas as pd

df =pd.read_csv("./vidoes_data__20250918_160244.csv")
df["date_created"] = pd.to_datetime(df["date_created"])
df["date_updated"] = pd.to_datetime(df["date_updated"])
df = df.set_index("date_updated")


df.head(20)
```

```python



trend_30s = (
    df.groupby( "description").resample("30sec")
      .agg({
          "video_id": "count",           # number of videos uploaded
          "total_views": "sum",          # total views
          "total_likes": "sum",
          "total_dislikes": "sum",
      })
      .rename(columns={"video_id": "video_count"})
)

trend_30s.head(30)
# df.groupby("video_id") \
#     .agg(
#             video_count=("total_likes", "max")
#             video_count=("total_views", "max")

#         )
```

```python
channel_stats = (
    df.groupby("channel_id") \
        .agg(
          total_videos=("video_id", "count"),
          total_views=("total_views", "sum"),
          avg_views=("total_views", "mean"),
          avg_likes=("total_likes", "mean"),
          avg_dislikes=("total_dislikes", "mean")
      )
)

channel_stats.head(30)
```
